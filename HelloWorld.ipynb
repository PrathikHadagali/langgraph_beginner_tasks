# Hello World Agent
### First LangGraph Graph!

```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph
```

```python
class AgentState(TypedDict):  #Our state schema
    message : str

def greeting_node(state: AgentState) -> AgentState:
    """Simple node that adds a greeting message to the state."""

    state['message'] = "Hey " + state["message"] + " ,How is your day going?"
    return state
```

### To create Graph in LangGraph we use StateGraph()



```python
graph = StateGraph(AgentState)


# How to add node --> we use inbuilt function "add_node()"
graph.add_node("greeter", greeting_node)


# To add START and END node as mentioned....--> call inbuilt function , "set_entry_point()"
graph.set_entry_point("greeter")
graph.set_finish_point("greeter")


# using inbuilt ".compile()", to compile the graph
app = graph.compile()

```

```python
from IPython.display import Image, display
display(Image(app.get_graph().draw_mermaid_png()))
```

# To Run we use .invoke()

```python
result = app.invoke({"message":"Bob"})
```

```python
result["message"]
```

# Exercise for Graph 1


```python
from typing import Dict, TypedDict
from langgraph.graph import StateGraph

class AgentState2(TypedDict):
    name : str
    message : str

def compliment_agent(state: AgentState2) -> AgentState2:
    state["message"] = state['name'] + ", you're doing an amazing job learning LangGraph!"
    return state

graph = StateGraph(AgentState2)
graph.add_node("compliment", compliment_agent)
graph.set_entry_point("compliment")
graph.set_finish_point("compliment")

app2 = graph.compile()

result2 = app2.invoke({"name":"Bob"})
result2["message"]
```
